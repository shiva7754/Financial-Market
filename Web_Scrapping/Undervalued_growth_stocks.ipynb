```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://finance.yahoo.com/screener/predefined/undervalued_growth_stocks/?count=100&offset=0"
```

```python
driver = webdriver.Chrome()
driver.get(url)
time.sleep(5)
```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
driver.quit()
```

```python
table = soup.find('table' , class_ = "W(100%)")
```

```python
table
```

```python
headers = []
thead = table.find('thead')
tr = thead.find('tr')
for th in tr.find_all('th'):
    text = th.get_text(strip=True)
    headers.append(text)
headers
```

```python
rows = []
tbody = table.find('tbody')
trows = tbody.find_all('tr')
for tr in trows:
    row = []
    for td in tr.find_all('td'):
        row_text = td.get_text(strip=True)
        row.append(row_text)
    rows.append(row)
```

```python
rows
```

```python
df = pd.DataFrame(rows , columns = headers)
df
```

```python
Undervalued_growth_stocks = df.drop('52 Week Range' ,axis=1)
```

```python
Undervalued_growth_stocks.shape
```

```python
Undervalued_growth_stocks.to_csv("Undervalued_growth_stocks.csv",index=False)
```
