```python
import pandas as pd
```

```python
Top_Gainers = pd.read_csv(r'E:\DA\Financial Market\Fetched_data\Top_Gainers.csv')
Most_active = pd.read_csv(r'E:\DA\Financial Market\Fetched_data\Most_active.csv')
Top_Losers = pd.read_csv(r'E:\DA\Financial Market\Fetched_data\Top_Losers.csv')
Undervalued_growth_stocks = pd.read_csv(r'E:\DA\Financial Market\Fetched_data\Undervalued_growth_stocks.csv')
```

```python
Top_Losers['Category'] = "Top_Losers"
Top_Losers
```

# Conacatinating the tables after adding category column to each table

```python
# Concatenate the dataframes
df_combined = pd.concat([Top_Gainers, Top_Losers, Most_active, Undervalued_growth_stocks])

# Reset the index (optional)
df_combined.reset_index(drop=True, inplace=True)
df_combined
```

# Counting Duplicates

```python
duplicate_count = df_combined.duplicated(subset='Symbol').sum()
duplicate_count
```

# marge categories if there are duplicates in table like if there are three same rows will add there categories in 1 row

```python
def merge_categories(group):
    combined_category = ', '.join(group['Category'].unique())
    group = group.iloc[0]  # Take the first occurrence of the duplicate rows
    group['Category'] = combined_category
    return group

# Apply the function to the dataframe grouped by 'Symbol' and 'Name'
df_final = df_combined.groupby(['Symbol']).apply(merge_categories).reset_index(drop=True)

# Display the final dataframe
df_final
```
